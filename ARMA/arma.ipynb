The ARMA model (Auto-Regressive + Moving Average) relates a signal $y_t$ to a random signal $\varepsilon_t$ as follows:
$$\displaystyle y_t = a_1 y_{t-1} + a_2 y_{t-2} + \cdots + \varepsilon_t + m_1 \varepsilon_{t-1} + m_2 \varepsilon_{t-2} + \cdots$$
OR
$$\displaystyle y_t = \sum_{i=1}^{N_{AR}-1} a_i y_{t-i} + \varepsilon_t + \sum_{j=1}^{N_{MA}-1} m_j \varepsilon_{t-j}$$

This can be re-expressed as:
$$\displaystyle a_0 y_{t-0} - \sum_{i=1}^{N_{AR}-1} a_i y_{t-i} = \sum_{j=0}^{N_{MA}-1} m_j \varepsilon_{t-j}$$
where $a_0=m_0=1$.

Notice the negative before $a_i$ terms, and the code below sets the corresponding values to negative.

# Generate some fake data

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_process import arma_generate_sample

# Change these values
include_AR = True
include_MA = False

AR, MA = [1], [1]
if include_AR:
    AR += [-0.5, -0.4]
if include_MA:
    MA += [0.5, 1.2, 2]
    
print('AR:', AR)
print('MA:', MA)

y = arma_generate_sample(AR, MA, nsample=1000, scale=0.1)
y -= y.mean()

plt.plot(y)
plt.show()
```

# Import real New York load data

From Nov. 30, 2020 (https://www.nyiso.com/load-data)

```python
import pandas as pd

data = pd.read_csv('load_ny.csv')
data.head()
```

```python
import random

# Select random identifier (PTID)
idx = random.choice(data.PTID)

# Extract full time-series of that PTID
y = data.Load[data.PTID == idx]

plt.plot(y)
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 4))
plot_acf (y, lags=20, zero=False, ax=ax[0])
plot_pacf(y, lags=20, zero=False, ax=ax[1])
plt.show()
```

### Observation:
Exhibits behavior of auto-regressive (AR) model.

### How about differential data?

Let $x_t = y_t - y_{t-1}$

```python
x = np.diff(y)

plt.plot(x)
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 4))
plot_acf (x, lags=20, zero=False, ax=ax[0])
plot_pacf(x, lags=20, zero=False, ax=ax[1])
plt.show()
```
