# Read csv file

Data source: [Local oscillations (eecs.utk.edu)](http://web.eecs.utk.edu/~kaisun/Oscillation/actualcases.html)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random

# Read file and indicate which rows are headers (non-numerical data)
data = pd.read_csv('ISO-NE_case5.csv', header=[0, 1, 2, 3])
data
```

# Extract raw voltage magnitudes

```python
t = data.filter(like='T')
VM = data.filter(like='VM')

fig = plt.plot(t, VM)
```

# Removing signals with missing data

Example of missing data

```python
column = VM.keys()[0]
plt.plot(t, VM[column])
plt.xlim(150, 160)
plt.show()
```

```python
print('Indices of null values BEFORE cleaning:', np.where(VM.isnull().values.any(axis=0))[0])

VM = VM.dropna(axis='columns')

print('Indices of null values AFTER cleaning:', np.where(VM.isnull().values.any(axis=0))[0])
```

# Detrend Data

To detrend is to subtract the mean

```python
X = VM - VM.mean()
fig = plt.plot(t, X)
```

# Scatter some measurements

```python
idx_measurement = [5, 15, 23]

subplot_idx = np.arange(9).reshape(3,3) + 1
plt.figure(figsize=(8,8))
for i, idx_i in enumerate(idx_measurement):
    for j, idx_j in enumerate(idx_measurement):
        plt.subplot(3, 3, subplot_idx[i,j])
        plt.plot(X.values[:,idx_i], X.values[:,idx_j], 'k.', alpha=0.01)
        if j == 0: plt.ylabel(f'Measurement {idx_i}')
        if i == 2: plt.xlabel(f'Measurement {idx_j}')
        plt.xticks([])
        plt.yticks([])
```

# Perform SVD (Singular Value Decomposition) & prove it worked

```python
# First, reshape matrix X such that:
# Number of columns: number of measurements per time-stamp
# Number of rows:    number of time points

X = X.values.T

X.shape
```

Data matrix $X$ can be decomposed via SVD as follows:
$$X = USV^H$$

```python
U, s, VH = np.linalg.svd(X)
dimU, dimV = U.shape[0], VH.shape[0]
S = np.zeros((dimU, dimV))
S[:dimU, :dimU] = np.diag(s)
print('U:', U.shape)
print('S:', S.shape)
print('V:', VH.T.shape)

X_svd = U @ S @ VH
plt.plot((X - X_svd).T)
plt.title('$X - U\Sigma V^H = 0$ (See y-axis order)')
plt.show()
```

# Show Covariance matrices of $X$ & $Y$, where $Y = PX$

```python
P = U.T
Y = P @ X
n = X.shape[1] # number of columns (measurements)
CovX = X @ X.T / (n-1)
CovY = Y @ Y.T / (n-1)

cmap = 'Greys'
plt.subplot(1, 2, 1);plt.imshow(CovX, cmap=cmap);plt.title('$\Sigma_X$', fontsize=16)
plt.subplot(1, 2, 2);plt.imshow(CovY, cmap=cmap);plt.title('$\Sigma_Y$', fontsize=16)
plt.show()
```

# Pick out top principal components

```python
diagCovY = np.diag(CovY)
percent_energy = diagCovY / np.sum(diagCovY) * 100

plt.figure(figsize=(16,3))
idx = range(diagCovY.shape[0])
plt.bar(idx, percent_energy, color='grey', alpha=0.5)

plt.title('Energy per principal component')
plt.xlabel('Principal Component')
plt.xticks([i for i in idx if not i % 2])
plt.ylabel('% of energy')
plt.show()
```

```python
plt.figure(figsize=(16,3))
plt.bar(idx, np.cumsum(percent_energy), color='grey', alpha=0.5)

# Plot 95% and 99% thresholds
plt.plot(idx, [95 for _ in idx], '--g', lw=3)
plt.plot(idx, [99 for _ in idx], '--r', lw=3)

plt.title('Cumulative Energy per principal component')
plt.xlabel('Principal Component')
plt.xticks([i for i in idx if not i % 2])
plt.ylabel('% of energy')
plt.ylim(80, 100)
plt.show()
```

# Show time series of top principal components

```python
p = 4
Yp = Y[:p]
plt.plot(Yp.T)
plt.xlabel('Time (s)')
plt.title(f'Top {p} principal components')
plt.show()
```

# Reconstruct $X$ from $Y_p$

```python
# Reconstruct using first p components
p = 4
Yp = Y[:p]
Xp = P[:p].T @ Yp

plt.figure(figsize=(16,4))
some_idx = random.choice(range(X.shape[0]))
plt.plot(X[some_idx], 'k', label='Original')
plt.plot(Xp[some_idx],'r', label='Reconstructed')
plt.title('Reconstruction from low-dimensional data')
plt.legend()
plt.xlim(0, 1000)
plt.show()
```
