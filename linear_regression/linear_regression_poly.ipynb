# Generate noisy data

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate clean data
x = np.linspace(-4, 4, 100)
y = 0.3*x**2 + 0.5*x - 1

# Add measurement noise
noise = 0.7 * np.random.randn(*y.shape)
y = y + noise

# Plot noisy data
plt.plot(x, y, 'ro', label='Noisy Data')
plt.show()
```

# Perform linear regression to estimate model parameters

$y=c_2 x^2 + c_1 x + c_0$  

OR  

$y = [x^2,~x,~1]\theta$  
where  
$\theta = [c_2,~c_1,~c_0]^T$

```python
theta = np.polyfit(x, y, 2)
c2, c1, c0 = np.round(theta, 3)
print(f'c2 = {c2}, c1 = {c1}, c0 = {c0}')
```

# Evaluate model over entire domain

```python
yhat = np.polyval(theta, x)

plt.plot(x, y, 'ro', label='Noisy Data')
plt.plot(x, yhat, 'g', lw=4, label='Estimated Fit')

# Show plot with labels
plt.legend(loc='best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
